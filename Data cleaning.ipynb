```python
# importing pandas library from python to work with messy data
import pandas as pd
```

```python
#importing the excel file as a pandas dataframe with column headers 'Col1' and 'Col2'
df = pd.read_excel('/home/sreelal/Downloads/input.xlsx',header=None, names=['col1','col2'])
```

```python
# it shows the no: of rows and columns in the dataframe
df.shape
```

```python
# showing a brief summary of the dataframe
df.info()
```

```python
# dropping the rows with null values
df1 = df.dropna()
```

```python
# printing the dataframe without null values
df1
```

```python
# showing dimensions of the new dataframe
df1.shape
```

```python
# default expression to increase the column width of dataframes
pd.set_option("max_colwidth",150)
```

```python
df1
```

```python
#dropping index to process data easily without index dependency 
df1.reset_index(drop=True, inplace=True)
```

```python
#Extracting rows with only Address from col1 in df1 and creating new dataframe df2 with only addresses
df2 = df1[df1['col1'].str.contains('Address')]

#Renaming column header to 'Address'
df2.rename(columns={'col2':'Address'},inplace=True)

#dropping index of df2
df2.reset_index(drop=True, inplace=True)
df2
```

```python
#Extracting rows with only store name from col1 in df1 and creating df3 with store names
df3 = df1[df1['col1'].str.contains("Store Name")]
#renaming column header to 'Store Name'
df3.rename(columns={'col2':'Store Name'},inplace=True)

#dropping index
df3.reset_index(drop=True, inplace=True)
df3
```

```python
#merging the two dataframes df2 and df3 to create df4 with addresses and store names as separate columns
df4 = pd.concat([df2,df3],axis = 1)
df4
```

```python
#extracting only required columns ie 'Address' and 'Store Name' as a new dataframe df5
df5 = df4[['Address','Store Name']]
df5
```

```python
# Now we have the cleaned data. So proceeding further to extract city and pincode from the address column
# To do that, first converting the entire address column to lowercase. Since there are adresses in which 
# the city name start with uppercase and lowercase.
df5.Address = df5.Address.apply(lambda x:x.lower())
```

```python
df5
```

```python
#To insert city name, creating a new column to the above dataframe with Nil as values
df5['City'] = 'Nil'
df5
```

```python

# To extract city name, we must have a list of cities whose data are copied from the site. So if the city name is
# found in the address, below code will extract the city name and insert into the column 'City' created above
for i in ['chennai','mumbai','bangalore','bengaluru','pune','hyderabad']:
    df5['City'][df5['Address'].str.contains(i)] = i
df5
```

```python
#To see all the stores in a particular city, for eg: taking bangalore
df5[df5['City'] == 'bangalore']
```

```python
# number of stores in pune city is 9
df5[df5['City'] == 'pune'].shape
```

```python
# number of stores in mumbai is 20
df5[df5['City'] == 'mumbai'].shape
```

```python
# number of addresses in which city name is not mentioned is 27
df5[df5['City'] == 'Nil'].shape
```

```python
# creating a new column pincode to extract pincodes from addresses
df5['Pincode'] = 'Nil'
df5
```

```python
#below code will extract all numbers from the address column and insert it into a list named num
num = []
import re
for i in df4.index:
    a = df5['Address'][i]
    b = re.findall('[0-9]+',a)
    num.append(b)
```

```python
num
```

```python
# from the list num, we need only pincodes ie numbers with 6 digits.So below code wil extract numbers with 6 digits
# and store it in list named pinocodes
pincodes = []
for j in range(len(num)):
    for k in range(len(num[j])):
        p = len(num[j][k])
        if p == 6:
            pincodes.append(num[j][k])
        else:
            pass
```

```python
pincodes
```

```python
# total number of pincodes available in the dataset.
len(pincodes)
```

```python
# So like we did for cities, now we have a list of pincodes in the dataset. 
# Below code will check in each address for each pincode from the list and if it is matching, it will insert the 
# pincode to the column 'Pincode'.
for i in pincodes:
    df5['Pincode'][df5['Address'].str.contains(i)] = i
df5
```

```python
# Now its time to write it back to an excel file with name output
df5.to_excel("output.xlsx",sheet_name='stores', index= False)
```
